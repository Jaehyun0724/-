###**음성** 인공지능의 활용 

주변에서 말하는 인공지능이 뭐가 있을까? 라고 생각했을때

가장 생각나는것은 휴대폰 플랫폼중 _ 시리, 빅스비, 구글 _ 등이 있다.

이러한 플랫폼은 마치 사람과 이야기 하는 것 처럼 느껴지며 

최근에 2020년 4월달 AI가 결합된 음성인식 기술이 단순히 음성을 알아듣고

텍스트로 변환해 주는 단계를 지나, 사람의 목소리 톤, 단어 등을 분석해

감정을 이해하는 '지능혁 컨텍센터'에 한 발짝 더 다가서고 있다고 한다.

이러한 장점을 통해 인공지능(AI)가 고객과 상담, 대응이 가능하다고 본다.

그 외에 네비게이션, 주문현황, 홈메이드 스피커 등 

우리 일상생활에 많은 도움을 주고 있다.

하지만 사투리경우 인공지능이 얼마나 잡아줄지 걱정이 된다.

###**이미지** 인공지능의 활용

사진을 찍을떄 시간을 맞춰서 타이밍 맞게 찍는게 불편한 감이 있다.

여기서 인공지능이 사람의 행동을 인식하게 된다면 어떤일이 발생할까?

그저 손가락 튕기는 것 만으로도 사진이 찍히게 된다.

번거로운 시간 제약이 사라진것 만으로도 많은 일들을 알수가 있는데

공항 보안검색대, cctv, 의료, 색구별, 이미지 분류등 여러가지가 있다.

하지만 그중 제일 발전이 되었으면 하는 것중 하나는 인식이다.

사람의 손, 얼굴, 사물, 동물, 날씨등 움직임을 포착하는 것 만으로

여러가지 행동을 예측할수가 있다.

이러한 행동 예측을 통해 실용성이 있다고 하면 

자연재해를 미리 알수가 있으며

이번 코로나 감연 사태를 기반해 사람들의 동선을 파악하기 쉬워진다.


또한 자동차 자율주행을 예를 들수가 있다. 


<hr/>

###**자율주행** 인공지능의 활용

자율주행이라면 자동차, 기차, 비행기 등 여러가지가 있는데

우리 주변에 가장 많은 자동차를 예를 들어보자

[YOUTUBE](https://www.youtube.com/watch?v=h4PuYA_ECwg)

위 동영상을 감상하면 운전자가 자리에 없어도 

자동으로 주차하는 모습이 보인다.

이러한 활용성으로 자동차 앞에 지나다니는 사람, 물체등의

움직임을 파악, 분석등을 통해 그냥 가도 될지 멈춰야 하는지

판단이 가능하다.

이처럼 자율주행은 과제에 배운 머신러닝 이라고 할수가 있다.

여기서 핵심은 스스로 멈춰야 하는지 가도 되는지 판단하는

학습이 된다는 점이다.

여기서 자율주행 자동차가 스스로 주행 하기 위해서는

 - 주변 환경 인지
 - 주행 도로 판단과 선정
 - 안전한 기능 제어

이러한 세 가지 기능이 반드시 수반돼야 한다.

그리고 딥러닝 기술 또한 중요한데

자동차의 인지, 판단, 제어 기능 향상에

딥러닝을 적용하는 연구가 활봘하게 진행되고 있다.

###**언어** 인공지능의 활용

불과 몇년전에는 손으로 일하는 때가 있었다.

지금은 키보드로 글씨를 입력하여 출력 하는 형식이지만 

인공지능(AI)가 개입하기 시작한다.

곧 다가올 인공지능이 어떠한 도움을 가지고 올까? 라고 봤을떄

지금 어떠한 활용들이 있는지 보면 -번역- 이라는 시스템이 존재하는데 

막상 번역해 보면 매끄로운 문장이 번역이 안된다는 것을 알수가 있다.

심지어 오역까지 나올수 있는데 이러한 문제점을 방안하기 위해 

빅데이터 자료를 사용한다. (빅데이터는 방대한 자료라고 생각하면 된다.)

여기서 더해 빅데이터로 단순한 번역, 통역이 아닌

인공지능이 감정까지 잡아 낼수 있다.

어떻게 감정을 잡아 낸다면 앞서 나온 음성, 이미지, 언어등의 빅데이터를 통해

사람의 톤, 언어, 표정 등을 분석하여 감정을 잡아 낼수 있다고 한다.

이러한 이야기는 곧 머지않아 실현된다고 믿는다.
